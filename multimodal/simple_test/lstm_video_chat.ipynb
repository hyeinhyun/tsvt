# lstm

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
#build model
model = Sequential()

model.add(LSTM(23,input_shape=(23,129), return_sequences=True))
model.add(LSTM(23,input_shape=(23,129), return_sequences=True))
model.add(LSTM(23,input_shape=(23,129)))
model.add(Dense(23))
model.add(Dense(2))
```

```python
from keras.callbacks import ModelCheckpoint
# Compile model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])

checkpoint = ModelCheckpoint(filepath='/home/ubuntu/gangmin/tsvt/multimodal/lstm/video_chat/'+'{epoch:02d}-{val_loss:.4f}.hdf5',monitor = 'val_loss',verbose=1,save_best_only=True)
```

```python
history = model.fit(X_train,y_train,epochs=100,batch_size=128,verbose=1,validation_data=[X_val,y_val],callbacks=[checkpoint])
```

```python
history_dict = history.history
import matplotlib.pyplot as plt
```
