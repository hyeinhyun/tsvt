```python
import pickle
with open('/home/ubuntu/gangmin/tsvt/multimodal/data/audio_energy_1.pickle','rb') as f:
    audio = pickle.load(f)
```

```python
import numpy as np
```

```python
p={}
for key in audio.keys():
    maxP = max(np.abs(audio[key]))
    p[key] = list(map(lambda x: np.abs(x) / maxP, audio[key]))
```

```python
p[key]
```

```python
for key in audio.keys():
    p[key] = list(map(lambda x: x*np.log(x) if x!=0 else 0 ,p[key]))
```

```python
p[key]
```

```python
H={}
for key in audio.keys():
    temp=[]
    for i,v in enumerate(p[key]):
        temp.append(-np.sum(p[key][max(0,i-int(23/2)):min(len(p[key]),i+int(23/2))]))
    H[key] = temp
```

```python
import pickle
with open('/home/ubuntu/gangmin/tsvt/multimodal/data/audio_H.pickle','wb') as f:
    pickle.dump(H,f)
```
